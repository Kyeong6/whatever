```python
pred = pred.drop(['label'], axis=1)
```

```python
# 데이터 확인
raw
```

```python
pred['time'] = pd.to_datetime(pred['time'], format='%Y-%m-%d %H:%M')
print(pred['time'].dtype)
```

```python
raw['datetime'] = pd.to_datetime(raw['datetime'], format='%Y-%m-%d %H:%M:%S')
print(raw['datetime'].dtype)
```

```python
# compared
compared = go.Figure()
```

```python
# Actual values
compared.add_trace(go.Scatter(x=raw['datetime'], y=raw['value'], mode='lines', name='Actual'))

# Predicted values
compared.add_trace(go.Scatter(x=pred['time'], y=pred['value'], mode='lines', name='Prediction'))

# Update x-axis to show weekly ticks
compared.update_layout(
    xaxis=dict(
        tickmode='linear',
        dtick=6 * 30 * 24 * 60 * 60 * 1000,
        tickformat='%Y-%m-%d'
    ),
    title='Actual vs Prediction',
    xaxis_title='Date',
    yaxis_title='Values',
    template='plotly_white'
)

# Show the plot
compared.show()

# Save the plot to HTML
pyo.plot(compared, filename='/Users/kyeong6/desktop/visualzation.html',
         auto_open=False)
```
